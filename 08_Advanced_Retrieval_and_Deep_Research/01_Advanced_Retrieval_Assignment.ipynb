## Task 3: Setting up QDrant!

Now that we have our documents, let's create a QDrant VectorStore with the collection name "investments_handbook".

We'll leverage OpenAI's [`text-embedding-3-small`](https://openai.com/blog/new-embedding-models-and-api-updates) because it's a very powerful (and low-cost) embedding model.

## Task 4: Naive RAG Chain

Since we're focusing on the "R" in RAG today - we'll create our Retriever first.

### R - Retrieval

This naive retriever will simply look at each review as a document, and use cosine-similarity to fetch the 10 most relevant documents.

> NOTE: We're choosing `10` as our `k` here to provide enough documents for our reranking process later

```python
naive_retriever = vectorstore.as_retriever(search_kwargs={"k" : 10})
```

### A - Augmented

We're going to go with a standard prompt for our simple RAG chain today! Nothing fancy here, we want this to mostly be about the Retrieval process.

```python
from langchain_core.prompts import ChatPromptTemplate

RAG_TEMPLATE = """\
You are a helpful and kind assistant. Use the context provided below to answer the question.

If you do not know the answer, or are unsure, say you don't know.

Query:
{question}

Context:
{context}
"""

rag_prompt = ChatPromptTemplate.from_template(RAG_TEMPLATE)
```

### G - Generation

We're going to leverage `gpt-4.1-nano` as our LLM today, as - again - we want this to largely be about the Retrieval process.

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(model="gpt-4.1-nano")
```

### LCEL RAG Chain

We're going to use LCEL to construct our chain.

> NOTE: This chain will be exactly the same across the various examples with the exception of our Retriever!

```python
from langchain_core.runnables import RunnablePassthrough
from operator import itemgetter
from langchain_core.output_parsers import StrOutputParser

naive_retrieval_chain = (
    # INVOKE CHAIN WITH: {"question" : "<<SOME USER QUESTION>>"}
    # "question" : populated by getting the value of the "question" key
    # "context"  : populated by getting the value of the "question" key and chaining it into the base_retriever
    {"context": itemgetter("question") | naive_retriever, "question": itemgetter("question")}
    # "context"  : is assigned to a RunnablePassthrough object (will not be called or considered in the next step)
    #              by getting the value of the "context" key from the previous step
    | RunnablePassthrough.assign(context=itemgetter("context"))
    # "response" : the "context" and "question" values are used to format our prompt object and then piped
    #              into the LLM and stored in a key called "response"
    # "context"  : populated by getting the value of the "context" key from the previous step
    | {"response": rag_prompt | chat_model, "context": itemgetter("context")}
)
```

## Task 5: Best-Matching 25 (BM25) Retriever

Taking a step back in time - [BM25](https://www.nowpublishers.com/article/Details/INR-019) is based on [Bag-Of-Words](https://en.wikipedia.org/wiki/Bag-of-words_model) which is a sparse representation of text.

### Question #1:

Give an example query where BM25 is better than embeddings and justify your answer.

##### Answer:

**Query:** *"What is an ILW in reinsurance?"*

BM25 would outperform embeddings here because:

1. **Exact keyword matching on rare acronyms** — "ILW" (Industry Loss Warranty) is a niche technical term. BM25 does direct lexical matching, so it will find any chunk containing the exact string "ILW" and rank it highly due to its rarity (high IDF score — rare terms get disproportionately high weight in BM25).

2. **Embeddings dilute rare terms** — Dense embedding models map tokens into a general-purpose semantic space. A rare domain acronym like "ILW" may not have a strong learned representation, so the embedding might drift toward generically similar concepts (e.g., "insurance" or "warranty") and retrieve less precise chunks.

The general principle: BM25 wins when the query hinges on **specific, rare, or domain-specific keywords** — acronyms, ticker symbols, regulation names like "Reg D", fund identifiers — where exact lexical overlap is more informative than semantic similarity.

## Task 6: Contextual Compression (Using Reranking)

Contextual Compression is a fairly straightforward idea: We want to "compress" our retrieved context into just the most useful bits.

There are a few ways we can achieve this - but we're going to look at a specific example called reranking.

The basic idea here is this:

- We retrieve lots of documents that are very likely related to our query vector
- We "compress" those documents into a smaller set of *more* related documents using a reranking algorithm.

We'll be leveraging Cohere's Rerank model for our reranker today!

All we need to do is the following:

- Create a basic retriever
- Create a compressor (reranker, in this case)

That's it!

Let's see it in the code below!

```python
from langchain.retrievers.contextual_compression import ContextualCompressionRetriever
from langchain_cohere import CohereRerank

compressor = CohereRerank(model="rerank-v3.5")
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=naive_retriever
)
```

Let's create our chain again, and see how this does!

```python
contextual_compression_retrieval_chain = (
    {"context": itemgetter("question") | compression_retriever, "question": itemgetter("question")}
    | RunnablePassthrough.assign(context=itemgetter("context"))
    | {"response": rag_prompt | chat_model, "context": itemgetter("context")}
)
```

```python
contextual_compression_retrieval_chain.invoke({"question" : "What are the key due diligence steps for alternative investments?"})["response"].content
```

### Question #2:

Explain how generating multiple reformulations of a user query can improve recall.

##### Answer:

A single query only captures one phrasing of the user's intent. Relevant documents may use different terminology, synonyms, or frame the concept from a different angle — so they won't match well against that one query vector.

Multi-query retrieval improves recall by:

1. **Covering lexical/semantic gaps** — The LLM generates reformulations that use different words for the same concept. E.g., the original query *"How does reinsurance diversify a portfolio?"* might produce reformulations like *"What are the correlation benefits of reinsurance as an asset class?"* and *"Why do institutional investors add reinsurance to reduce portfolio risk?"*. Each phrasing is closer in embedding space to a different subset of relevant chunks.

2. **Capturing multiple facets** — A broad question often has sub-aspects. Reformulations can tease these apart, retrieving documents that address each facet individually.

3. **Union of results** — The retriever takes the **union of all unique documents** across all reformulated queries. Even if each individual query only retrieves some relevant chunks, the combined set has higher recall than any single query alone.

The trade-off is increased latency and cost (multiple LLM calls for reformulation + multiple retrieval passes), but the recall improvement is often substantial for complex or ambiguous queries.

## Task 8: Parent Document Retriever

A "small-to-big" strategy - the Parent Document Retriever works based on a simple strategy:

1. We split the full document into large "parent" chunks (e.g. 2000 characters).
2. Each parent chunk is further split into smaller "child" chunks (e.g. 400 characters).
3. The child chunks are stored in a VectorStore, while the parent chunks are stored in an in-memory docstore.
4. When we query our Retriever, we do a similarity search comparing our query vector to the child chunks.
5. Instead of returning the child chunks, we return their associated parent chunks.

The basic idea is:

- **Search** for small, focused chunks (better semantic matching)
- **Return** big chunks (richer surrounding context)

The intuition is that we're likely to find the most relevant information by limiting the amount of semantic information encoded in each embedding vector - but we're likely to miss relevant surrounding context if we only use that information.

Let's start by defining our parent and child splitters.

```python
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
from langchain_text_splitters import RecursiveCharacterTextSplitter
from qdrant_client import QdrantClient, models

parent_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)
child_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)
```

We'll need to set up a new QDrant vectorstore - and we'll use another useful pattern to do so!

> NOTE: We are manually defining our embedding dimension, you'll need to change this if you're using a different embedding model.

```python
from langchain_qdrant import QdrantVectorStore

client = QdrantClient(location=":memory:")

client.create_collection(
    collection_name="investments_parent_child",
    vectors_config=models.VectorParams(size=1536, distance=models.Distance.COSINE)
)

parent_document_vectorstore = QdrantVectorStore(
    collection_name="investments_parent_child", embedding=OpenAIEmbeddings(model="text-embedding-3-small"), client=client
)
```

Now we can create our `InMemoryStore` that will hold our "parent documents" - and build our retriever!

```python
store = InMemoryStore()

parent_document_retriever = ParentDocumentRetriever(
    vectorstore=parent_document_vectorstore,
    docstore=store,
    child_splitter=child_splitter,
    parent_splitter=parent_splitter,
)
```

By default, this is empty as we haven't added any documents - let's add some now!

```python
parent_document_retriever.add_documents(raw_docs, ids=None)
```

We'll create the same chain we did before - but substitute our new `parent_document_retriever`.

```python
parent_document_retrieval_chain = (
    {"context": itemgetter("question") | parent_document_retriever, "question": itemgetter("question")}
    | RunnablePassthrough.assign(context=itemgetter("context"))
    | {"response": rag_prompt | chat_model, "context": itemgetter("context")}
)
```

## Task 9: Ensemble Retriever

In brief, an Ensemble Retriever simply takes 2, or more, retrievers and combines their retrieved documents based on a rank-fusion algorithm.

In this case - we're using the [Reciprocal Rank Fusion](https://plg.uwaterloo.ca/~gvcormac/cormacksigir09-rrf.pdf) algorithm.

Setting it up is as easy as providing a list of our desired retrievers - and the weights for each retriever.

```python
from langchain.retrievers import EnsembleRetriever

retriever_list = [bm25_retriever, naive_retriever, parent_document_retriever, compression_retriever, multi_query_retriever]
equal_weighting = [1/len(retriever_list)] * len(retriever_list)

ensemble_retriever = EnsembleRetriever(
    retrievers=retriever_list, weights=equal_weighting
)
```

We'll pack *all* of these retrievers together in an ensemble.

```python
ensemble_retrieval_chain = (
    {"context": itemgetter("question") | ensemble_retriever, "question": itemgetter("question")}
    | RunnablePassthrough.assign(context=itemgetter("context"))
    | {"response": rag_prompt | chat_model, "context": itemgetter("context")}
)
```

Let's look at our results!

## Task 10: Semantic Chunking

While this is not a retrieval method - it *is* an effective way of increasing retrieval performance on corpora that have clean semantic breaks in them.

Essentially, Semantic Chunking is implemented by:

1. Embedding all sentences in the corpus.
2. Combining or splitting sequences of sentences based on their semantic similarity based on a number of [possible thresholding methods](https://python.langchain.com/docs/how_to/semantic-chunker/):
  - `percentile`
  - `standard_deviation`
  - `interquartile`
  - `gradient`
3. Each sequence of related sentences is kept as a document!

Let's see how to implement this!

We'll use the `percentile` thresholding method for this example which will:

Calculate all distances between sentences, and then break apart sequences of setences that exceed a given percentile among all distances.

```python
from langchain_experimental.text_splitter import SemanticChunker

semantic_chunker = SemanticChunker(
    embeddings,
    breakpoint_threshold_type="percentile"
)
```

Now we can split our documents.

```python
semantic_documents = semantic_chunker.split_documents(raw_docs)
```

Let's create a new vector store.

```python
semantic_vectorstore = QdrantVectorStore.from_documents(
    semantic_documents,
    embeddings,
    location=":memory:",
    collection_name="investments_handbook_semantic_chunks"
)
```

We'll use naive retrieval for this example.

```python
semantic_retriever = semantic_vectorstore.as_retriever(search_kwargs={"k" : 10})
```

Finally we can create our classic chain!

```python
semantic_retrieval_chain = (
    {"context": itemgetter("question") | semantic_retriever, "question": itemgetter("question")}
    | RunnablePassthrough.assign(context=itemgetter("context"))
    | {"response": rag_prompt | chat_model, "context": itemgetter("context")}
)
```

And view the results!

```python
semantic_retrieval_chain.invoke({"question" : "How does reinsurance provide portfolio diversification benefits?"})["response"].content
```

### Question #3:

If sentences are short and highly repetitive (e.g., FAQs), how might semantic chunking behave, and how would you adjust the algorithm?

##### Answer:

**The problem:** Semantic chunking computes embedding distances between consecutive sentences and splits where the distance exceeds a threshold. With short, repetitive sentences (like FAQs), the embeddings will be very similar to each other — the distances between consecutive sentences will be uniformly low. This means:

- **Few or no breakpoints get triggered** — The algorithm sees little semantic variation, so it lumps most or all sentences into one massive chunk, defeating the purpose of chunking.
- **Or with percentile thresholding**, it still must split at *some* percentile of distances, so it may split at arbitrary, meaningless points where the tiny distance differences are just noise.

**How to adjust:**

1. **Switch thresholding method** — Use `standard_deviation` or `gradient` instead of `percentile`. Gradient-based detection looks for *rate of change* in distances, which can catch subtle topic shifts even when absolute distances are small.

2. **Lower the breakpoint threshold** — Make the algorithm more sensitive to smaller semantic shifts (e.g., lower the percentile cutoff from the default).

3. **Use a hybrid approach** — Combine semantic chunking with a max chunk size constraint, so even if the algorithm doesn't detect a semantic break, chunks are still bounded to a reasonable length.

4. **Pre-group by structure** — For FAQs specifically, split on structural cues first (e.g., question marks, numbered items, headers), then apply semantic chunking within each group.

```python
### Step 1: Generate a synthetic "golden dataset" using Ragas TestsetGenerator

from ragas.testset import TestsetGenerator
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

generator_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4.1-mini"))
generator_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings(model="text-embedding-3-small"))

generator = TestsetGenerator(llm=generator_llm, embedding_model=generator_embeddings)
testset = generator.generate_with_langchain_docs(raw_docs, testset_size=10)

testset_df = testset.to_pandas()
testset_df.head()
```

```python
### Step 2: Run each retriever on the synthetic queries and collect results

from ragas import SingleTurnSample, EvaluationDataset
import time

retriever_map = {
    "naive": naive_retriever,
    "bm25": bm25_retriever,
    "compression": compression_retriever,
    "multi_query": multi_query_retriever,
    "parent_document": parent_document_retriever,
    "ensemble": ensemble_retriever,
    "semantic": semantic_retriever,
}

# Retrievers that hit the Cohere reranker (Trial key: 10 calls/min)
cohere_retrievers = {"compression", "ensemble"}

all_results = {}

for name, retriever in retriever_map.items():
    print(f"\n--- Running retriever: {name} ---")
    samples = []
    start = time.time()

    for i, (_, row) in enumerate(testset_df.iterrows()):
        query = row["user_input"]
        reference = row["reference"]

        # Rate-limit Cohere retrievers to stay under 10 calls/min
        if name in cohere_retrievers and i > 0:
            time.sleep(7)

        # Retrieve docs once, then pass them directly to the prompt
        docs = retriever.invoke(query)
        contexts = [d.page_content for d in docs]

        # Generate response reusing already-retrieved context (no second retriever call)
        prompt_input = rag_prompt.invoke({"question": query, "context": "\n\n".join(contexts)})
        response = chat_model.invoke(prompt_input).content

        samples.append(SingleTurnSample(
            user_input=query,
            retrieved_contexts=contexts,
            response=response,
            reference=reference,
        ))

    elapsed = time.time() - start
    all_results[name] = {"samples": samples, "latency": elapsed}
    print(f"  {name}: {len(samples)} samples in {elapsed:.1f}s")
```

```python
### Step 3: Evaluate each retriever with Ragas retriever-specific metrics

from ragas import evaluate
from ragas.metrics import LLMContextRecall, Faithfulness, LLMContextPrecisionWithReference, ContextEntityRecall

evaluator_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4.1-mini"))

metrics = [
    LLMContextRecall(llm=evaluator_llm),
    LLMContextPrecisionWithReference(llm=evaluator_llm),
    ContextEntityRecall(llm=evaluator_llm),
    Faithfulness(llm=evaluator_llm),
]

eval_scores = {}

for name, data in all_results.items():
    print(f"\n--- Evaluating: {name} ---")
    dataset = EvaluationDataset(samples=data["samples"])
    result = evaluate(dataset=dataset, metrics=metrics)
    eval_scores[name] = {
        "context_recall": result["context_recall"],
        "context_precision": result["llm_context_precision_with_reference"],
        "entity_recall": result["context_entity_recall"],
        "faithfulness": result["faithfulness"],
        "latency_s": data["latency"],
    }
    print(f"  {eval_scores[name]}")
```

```python
### Step 4: Compare results in a table

import pandas as pd
import numpy as np

# Average the per-sample scores into single values per retriever
avg_scores = {}
for name, data in eval_scores.items():
    avg_scores[name] = {
        k: np.mean(v) if isinstance(v, list) else v
        for k, v in data.items()
    }

scores_df = pd.DataFrame(avg_scores).T
scores_df = scores_df.sort_values("context_recall", ascending=False)
scores_df.style.format("{:.4f}") \
    .highlight_max(axis=0, subset=["context_recall", "context_precision", "entity_recall", "faithfulness"]) \
    .highlight_min(axis=0, subset=["latency_s"])
```

Key observations:

- **Performance**: Multi-query achieves the highest context recall (0.98) by generating query reformulations that surface documents a single query would miss. However, parent_document nearly matches it (0.92) while dominating on precision (0.97 vs 0.75) — meaning almost every chunk it returns is relevant. BM25 performs worst across all metrics (0.42 recall, 0.44 precision), confirming that pure lexical matching struggles with this domain's varied terminology.

- **Cost**: BM25 and naive are cheapest (no extra API calls beyond embeddings). Multi-query adds LLM calls for query reformulation. Compression and ensemble add Cohere reranker calls — and notably, compression underperformed expectations (0.74 recall), possibly because reranking aggressively filtered out relevant chunks. Ensemble compounds the cost of all sub-retrievers without a clear performance gain over simpler methods.

- **Latency**: BM25 (38s) and naive (39s) are fastest. Parent_document (42s) is nearly as fast while delivering far better precision. Multi-query (60s) and semantic (65s) are moderate. Ensemble is the slowest (148s) — 3.5x slower than parent_document — due to running all five sub-retrievers, yet its recall (0.95) doesn't justify the cost over multi-query (0.98) or parent_document (0.92).

**Verdict**: For this dataset, **parent_document** is the best retriever — it achieves the highest precision and faithfulness, strong recall, and fast latency. Its "small-to-big" strategy (search on focused child chunks, return rich parent chunks) is well-suited to this handbook's structure where related information clusters in contiguous sections.
