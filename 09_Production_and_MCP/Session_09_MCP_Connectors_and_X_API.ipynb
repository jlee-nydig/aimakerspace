## Task 2: X API as LangChain Tools

Instead of relying on a community-built MCP server for X, we'll call the **X API v2** directly and wrap our functions with the `@tool` decorator. This makes them available to our LangGraph agent as callable tools.

This is a key architectural decision: **not everything needs to be an MCP server**. Wrapping a simple API call as a `@tool` is often simpler and more transparent.

**Documentation:**
- [LangChain Tools Conceptual Guide](https://python.langchain.com/docs/concepts/tools/)
- [X API v2 Documentation](https://developer.x.com/en/docs/x-api)

````python
import requests
import json
from langchain_core.tools import tool

BEARER_TOKEN = os.environ.get("X_BEARER_TOKEN")


@tool
def search_recent_posts(query: str, max_results: int = 20) -> str:
    """Search recent X/Twitter posts using the v2 API.
    Returns posts from the last 7 days matching the query.
    Use this for keyword searches, hashtag searches, or general topic searches.

    Args:
        query: The search query (e.g., 'Stone Ridge', '#bitcoin', 'reinsurance market')
        max_results: Number of results to return (10-100, default 20)
    """
    url = "https://api.x.com/2/tweets/search/recent"
    headers = {"Authorization": f"Bearer {BEARER_TOKEN}"}
    params = {
        "query": query,
        "max_results": min(max(max_results, 10), 100),
        "tweet.fields": "created_at,public_metrics,author_id,text",
        "expansions": "author_id",
        "user.fields": "name,username",
    }

    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    data = response.json()

    tweets = data.get("data", [])
    if not tweets:
        return "No posts found for this query."

    result_lines = [f"Found {len(tweets)} posts:\n"]
    for t in tweets:
        metrics = t.get("public_metrics", {})
        result_lines.append(
            f"[{t.get('created_at', 'unknown')[:10]}] "
            f"{t['text'][:200]}\n"
            f"  Likes: {metrics.get('like_count', 0)} | "
            f"Retweets: {metrics.get('retweet_count', 0)}"
        )
    return "\n\n".join(result_lines)


@tool
def get_user_posts(username: str, max_results: int = 20) -> str:
    """Get recent original posts (no retweets) from a specific X/Twitter user.
    Use this when you want to see what a specific account has been posting.

    Args:
        username: The X/Twitter handle without the @ sign (e.g., 'StoneRidgeAM')
        max_results: Number of results to return (10-100, default 20)
    """
    query = f"from:{username} -is:retweet"
    return search_recent_posts.invoke({"query": query, "max_results": max_results})


x_api_tools = [search_recent_posts, get_user_posts]
print(f"Created {len(x_api_tools)} X API tools: {[t.name for t in x_api_tools]}")
````

````python
from typing import Annotated, Literal
from typing_extensions import TypedDict

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

# Combine all tools: X API tools + GitHub MCP tools
all_tools = x_api_tools + github_mcp_tools
print(f"Total tools available to agent: {len(all_tools)}")
print(f"  X API tools: {[t.name for t in x_api_tools]}")
print(f"  GitHub MCP tools: {[t.name for t in github_mcp_tools]}")
````

````python
# Step 1: Define the Agent State
class AgentState(TypedDict):
    messages: Annotated[list, add_messages]


# Step 2: Define the system prompt
SYSTEM_PROMPT = """You are an AI investment research assistant that can search X/Twitter posts for \
market sentiment and manage GitHub repositories.

You have two categories of tools:
1. X API tools: search_recent_posts, get_user_posts \u2014 for searching and retrieving X/Twitter posts \
about investments, market sentiment, Stone Ridge, bitcoin, reinsurance, etc.
2. GitHub MCP tools: for creating repos, committing files, creating branches, opening PRs, etc.

When asked to analyze market sentiment, retrieve posts first using the X API tools, then provide \
a structured markdown summary with: Overview, Key Themes, Notable Posts, and Summary Statistics.

When asked to perform GitHub operations, use the appropriate GitHub MCP tool.
Always use the available tools when appropriate. Be concise in your responses."""


# Step 3: Bind tools to the LLM
llm_with_tools = llm.bind_tools(all_tools)


# Step 4: Define the agent node
def agent_node(state: AgentState):
    """The agent node \u2014 calls the LLM with the current conversation and available tools."""
    messages = [SystemMessage(content=SYSTEM_PROMPT)] + state["messages"]
    response = llm_with_tools.invoke(messages)
    return {"messages": [response]}


# Step 5: Define the tool node
tool_node = ToolNode(all_tools, handle_tool_errors=True)


# Step 6: Define routing logic
def should_continue(state: AgentState) -> Literal["tools", "end"]:
    """Determine whether to call tools or end the conversation."""
    last_message = state["messages"][-1]
    if hasattr(last_message, "tool_calls") and last_message.tool_calls:
        return "tools"
    return "end"


# Step 7: Build the graph
workflow = StateGraph(AgentState)

workflow.add_node("agent", agent_node)
workflow.add_node("tools", tool_node)

workflow.add_edge(START, "agent")
workflow.add_conditional_edges("agent", should_continue, {"tools": "tools", "end": END})
workflow.add_edge("tools", "agent")

# Compile with MemorySaver for short-term memory across turns
checkpointer = MemorySaver()
agent = workflow.compile(checkpointer=checkpointer)

print("Agent compiled with memory and tools!")
````

````python
from IPython.display import Image, display

display(Image(agent.get_graph().draw_mermaid_png()))
````

````python
# Use a consistent thread_id so the agent remembers context across all tasks
config = {"configurable": {"thread_id": "mcp-investment-workflow-1"}}


async def ask_agent(user_message: str) -> str:
    """Send a message to the agent and return its final response."""
    response = await agent.ainvoke(
        {"messages": [HumanMessage(content=user_message)]},
        config,
    )
    return response["messages"][-1].content
````

How could X/Twitter sentiment analysis complement traditional investment research methods like reading investor letters?

##### Answer

X/Twitter sentiment analysis complements traditional investment research in several key ways:

- **Real-time signal vs. periodic reporting**: Investor letters like Stone Ridge's are published quarterly or annually. X provides a continuous stream of market sentiment, capturing reactions to breaking news, earnings releases, and macro events as they happen — filling the gap between formal publications.

- **Breadth of perspectives**: An investor letter reflects a single firm's viewpoint. X aggregates opinions from retail investors, institutional analysts, journalists, and industry insiders simultaneously, surfacing consensus views or contrarian signals that a single letter cannot capture.

- **Leading indicator potential**: Social media sentiment can shift before formal research catches up. A surge in negative posts about a sector (e.g., reinsurance after a major hurricane) may signal repricing before it shows up in analyst reports, giving researchers an early-warning mechanism.

- **Validation and triangulation**: When Stone Ridge's letter argues for a thesis like bitcoin allocation or reinsurance as an asset class, X sentiment data can validate whether the broader market agrees, disagrees, or is simply unaware — providing context for how "consensus" or "contrarian" a position actually is.

- **Identifying blind spots**: Traditional research focuses on known themes. X surface discussions about emerging topics, regulatory shifts, or grassroots narratives (e.g., meme-driven interest in specific assets) that may not yet appear in institutional research but could affect market dynamics.

### ❓ Question #2

Compare using GitHub MCP tools (through a LangGraph agent) to traditional `git` commands. What felt easier? What felt harder or less transparent?

##### Answer

**What felt easier:**

- **Natural language intent**: Instead of remembering exact `git` and `gh` CLI syntax (e.g., `gh repo create --public --description "..." --clone`, `git checkout -b`, `gh pr create --base main --head add-metadata`), you describe what you want in plain English and the agent selects the right MCP tool and parameters. This lowers the barrier for complex multi-step workflows like "create a branch, commit a file, and open a PR."

- **Orchestration across steps**: The agent with memory chains operations together automatically. It remembers the summary it generated from X posts and commits it to GitHub without you having to copy/paste content between steps. With traditional CLI, you'd manage that state yourself.

- **Discoverability**: The agent knows all available GitHub operations from the MCP server's tool list. With `git`/`gh` CLI, you need to know which subcommand exists or search documentation.

**What felt harder or less transparent:**

- **Debugging failures**: When a `git push` fails, the CLI gives you the exact error and you fix it directly. When an MCP tool call fails inside the agent, the error is wrapped in the agent's response and you lose visibility into exactly what request was sent and what the API returned.

- **Predictability**: With CLI commands, you know exactly what will execute. The agent may choose unexpected tools or parameters — for example, it might pick a different branch name or commit message than what you intended if the prompt is ambiguous.

- **Latency**: Each GitHub operation requires an LLM round-trip (to decide which tool to call) plus the MCP server call. A direct `gh` CLI command executes in under a second; the agent workflow takes several seconds per operation due to LLM inference.

- **Auditability**: `git log` gives you a clear history of exactly what happened. With the agent, you'd need to trace through the LangSmith logs or conversation history to reconstruct the sequence of MCP tool calls that were made.

### ❓ Question #3

You used MCP for GitHub but wrapped the X API as a `@tool` directly. What are the tradeoffs of consuming an API through an MCP server versus wrapping it as a LangChain tool? When would each approach make more sense?

##### Answer

| | MCP Server | LangChain `@tool` |
|---|---|---|
| **Setup** | Requires a running server process (local or remote) with its own transport layer | Just a Python function with a decorator — no infrastructure |
| **Maintenance** | Server maintained by a third party (e.g., GitHub); updates happen independently | You own the code; API changes require you to update the wrapper |
| **Tool count** | One connection can expose dozens of tools automatically (GitHub MCP gave us 30+ tools) | Each endpoint needs a manually written function |
| **Portability** | MCP is protocol-standard — any MCP-compatible client can use it, not just LangChain | Tightly coupled to LangChain/LangGraph |
| **Transparency** | Tool behavior is a black box defined by the server | Full control over request construction, error handling, and response formatting |
| **Debugging** | Harder to inspect what the server sends to the upstream API | You can log/print every request and response directly |

**When MCP makes more sense:**
- The API is complex with many endpoints (like GitHub — repos, issues, PRs, branches, files, etc.) and an official MCP server already exists
- You want to stay current with API changes without maintaining wrappers yourself
- You need the tools to be reusable across different agent frameworks, not just LangChain

**When `@tool` makes more sense:**
- The API is simple with only a few endpoints (like X search — just 2-3 calls)
- You need fine-grained control over request parameters, error handling, or response formatting
- No reliable MCP server exists for the API, and building one would be over-engineering
- You want to add custom logic like rate limiting, caching, or response truncation that an MCP server doesn't provide

3. **Commit through the MCP workflow**: create branch, commit, open PR

````python
# --- Step 1: Retrieve posts about a second investment topic ---

result1 = await ask_agent(
    "Search X for recent posts about reinsurance market trends and catastrophe bonds."
)
print("=== Topic 1: Reinsurance & Cat Bonds ===")
print(result1[:800])

# --- Step 2: Retrieve posts about a contrasting topic ---

result2 = await ask_agent(
    "Now search X for recent posts about energy assets and oil market outlook."
)
print("\n=== Topic 2: Energy Assets & Oil ===")
print(result2[:800])

# --- Step 3: Generate a structured comparison as comparison.md ---

comparison = await ask_agent(
    "Compare the two sets of posts you just retrieved (reinsurance/cat bonds vs. energy/oil). "
    "Create a structured markdown document called comparison.md with these sections:\n"
    "- Title: 'Investment Sentiment Comparison'\n"
    "- Section 1: Reinsurance & Catastrophe Bonds — key themes, sentiment, notable posts\n"
    "- Section 2: Energy Assets & Oil — key themes, sentiment, notable posts\n"
    "- Section 3: Cross-Topic Analysis — similarities, differences, and investment implications\n"
    "- Section 4: Methodology — queries used, date range, number of posts\n"
    "Format it as ready-to-commit markdown."
)
print("\n=== Comparison Document ===")
print(comparison[:1000])

# --- Step 4: Commit through the MCP workflow ---

# 4a: Create a feature branch
branch_result = await ask_agent(
    "Create a new branch called `add-comparison` in my `investment-sentiment-analysis` repo "
    "from the `main` branch."
)
print("\n=== Branch Creation ===")
print(branch_result)

# 4b: Commit comparison.md to the branch
commit_result = await ask_agent(
    "Create a file called `comparison.md` on the `add-comparison` branch in my "
    "`investment-sentiment-analysis` repo. Use the comparison document you just generated "
    "as the file content. Commit message: 'Add cross-topic sentiment comparison'."
)
print("\n=== Commit ===")
print(commit_result)

# 4c: Open a pull request
pr_result = await ask_agent(
    "Open a pull request in my `investment-sentiment-analysis` repo from `add-comparison` "
    "to `main`. Title: 'Add cross-topic sentiment comparison'. Description should summarize "
    "that this PR adds a comparative analysis of reinsurance/cat bonds vs energy/oil "
    "market sentiment from X posts."
)
print("\n=== Pull Request ===")
print(pr_result)
````
